```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
'''# Sample movie dataset
data = {'Title': ['Movie 1', 'Movie 2', 'Movie 3', 'Movie 4'],
        'Description': ['Action movie with car chases and explosions',
                         'Romantic comedy with love triangles',
                         'Sci-fi adventure with aliens and space exploration',
                         'Animated film with talking animals']}'''
```

```python
df = pd.DataFrame(movie_dataset)
```

```python

# User's preferences
user_preferences = 'I love action movies with car chases and explosions'
```

```python
# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movie_dataset['keywords'])
```

```python
# TF-IDF Vectorizer for user preferences 
user_tfidf = tfidf_vectorizer.transform([user_preferences])
```

```python
# Calculate cosine similarity
cosine_sim = linear_kernel(user_tfidf, tfidf_matrix)
```

```python
# Get movie recommendations based on similarity
movie_indices = cosine_sim.argsort()[0][::-1]
recommended_movies = movie_dataset['original_title'].iloc[movie_indices]
```

```python
print("User Preferences:", user_preferences)
print("Recommended Movies:")
print(recommended_movies)
```

# K Nearest neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Sample data (replace this with your dataset)
X =  movie_dataset[['budget','genres','homepage','id','keywords','original_language','overview','popularity','production_companies','production_countries','release_date','revenue','runtime','spoken_languages','status','tagline','vote_average','vote_count']]
y = movie_dataset['title']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the classifier
knn.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = knn.predict(X_test)

```

```python
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
